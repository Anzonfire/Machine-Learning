```python
# libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#  plots
%matplotlib inline

# dataset
df1 = pd.read_csv("data.csv", encoding='ISO-8859-1')

# Inspect dataset
print("First 5 rows:")
print(df1.head())

print("Last 5 rows:")
print(df1.tail())

print("Dataset Info:")
print(df1.info())

print("Statistical Summary:")
print(df1.describe())

# Check missing values
print("Missing Values:")
print(df1.isnull().sum())


df1 = df1.dropna(subset=['Description'])

# Sales per product
df1['Sales'] = df1['Quantity'] * df1['UnitPrice']

sales_product = df1.groupby('Description')['Sales'].sum().sort_values(ascending=False).head(10)

# Bar Chart
plt.figure()
sales_product.plot(kind='bar')
plt.title("Top 10 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=90)
plt.show()

# Line Chart
plt.figure()
sales_product.plot()
plt.title("Sales Trend of Top Products")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=90)
plt.show()
```
